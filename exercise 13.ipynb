Crea un dataframe di dieci righe, quattro colonne con valori casuali. Scrivi un programma Pandas per evidenziare le colonne specifiche del dataframe con colori diversi.

```python
import pandas as pd
import numpy as np

Impiegati = pd.DataFrame({
    "nome" : ["Marta", "Marco", "Manuela", "Luca", "Carla", "Gianni", "Aldo", "Giulia", "Alba", "Cloe", "Valerio"],
    "ID" : [12, 56, 44, 45, 67, 75, 32, 54, 99, 64, 46],
    "età" : [25, 29, 30, 31, 24, 26, 21, 24, 32, 34, 28],
    "laurea": ["si", "no", "si", "si", "si", "no", "no", "si", "si", "no", "si"]})

Impiegati
```

```python

def highlight_cols(x):
    Impiegati = x.copy()
    Impiegati.loc[:,:] = 'background-color: red'
    Impiegati[['nome','età']] = 'background-color: green'
    
    return Impiegati   

Impiegati.style.apply(highlight_cols, axis=None)
```
