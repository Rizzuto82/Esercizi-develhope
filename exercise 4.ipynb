Scrivere un programma Panda per filtrare i record dal set di dati sul consumo mondiale di alcol in cui il consumo medio di bevande per persona (birra, liquore e vino combinati) rientra nell'intervallo compreso tra 0,5 e 2,5. Recupera i record che includono il paese, il consumo medio, la regione e il PIL pro capite.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\rizzu\\OneDrive\\Desktop\\visual studio\\esercizi piattaforma numpy pandas\\World_Alcohol_Dataset - MAIN.csv", na_values = "?")

print(df)
```

```python
df.fillna(0, inplace = True)
```

```python
df[["beer_servings", "spirit_servings", "wine_servings"]].mean() 
```

```python
df["consumo_medio"] = df[["beer_servings", "spirit_servings", "wine_servings"]].mean(axis=1)
df["consumo_medio"]
```

```python
df.head(3)
```

```python
dff = df[(df["consumo_medio"] >= 0.5) & (df["consumo_medio"] <= 2.5)]
dff
```
